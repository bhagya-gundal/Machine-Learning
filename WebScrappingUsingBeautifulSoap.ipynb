```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import requests
from bs4 import BeautifulSoup

plt.style.use('fivethirtyeight')

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

defining a function which can load data using url and creates a table

```python
def sport_data(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'lxml')
    table = soup.find_all("table")[0]
    table_head = soup.find_all("thead")[0]
    
    return pd.read_html(str(table))[0]
```

Defining the prior variables using uniform distribution

```python
n_draw = 2000
prior_ni = pd.Series(np.random.uniform(0, 1, size = n_draw)) 
plt.figure(figsize=(8,5))
plt.hist(prior_ni)
plt.title('Uniform distribution(0,1)')
plt.xlabel('Prior on AVG')
plt.ylabel('Frequency')
```

Building a Bayesian Model for finding the posterior paramters and it is done using binomial distribution

```python
def posterior(n_atbat, k_hits, prior):
    hit = list()
    for p in prior:
        hit.append(np.random.binomial(n_atbat, p))
        
    posterior = prior[list(map(lambda x: x == k_hits, hit))]
    plt.figure(figsize=(8,5))
    plt.hist(posterior)
    plt.title('Posterior distribution')
    plt.xlabel('Posterior on Mean')
    plt.ylabel('Frequency')
    print('Number of draws left: %d, Posterior mean: %.3f, Posterior median: %.3f, Posterior 95%% quantile interval: %.3f-%.3f' % 
      (len(posterior), posterior.mean(), posterior.median(), posterior.quantile(.025), posterior.quantile(.975)))

Rouwen_Hennings = int(Bundesliga_data[['GS','ASST']].iloc[3][0])
Francisco_Alcacer = int(Bundesliga_data[['GS','ASST']].iloc[2][0])
print(Rouwen_Hennings)
print(Francisco_Alcacer)
posterior(Rouwen_Hennings, Francisco_Alcacer, prior_ni)
```

Implementing beta distribution on prior variables 

```python
n_draw = 20000
prior_atbat = int(KB18.GP)
prior_hits = int(KB18.A)
prior_i = pd.Series(np.random.beta(prior_atbat + 1, prior_atbat - prior_hits + 1, size = n_draw)) 
plt.figure(figsize=(8,5)) 
plt.hist(prior_i)
plt.title('Beta distribution(a=%d, b=%d)' % (prior_hits + 1,prior_hits - prior_atbat + 1))
plt.xlabel('Prior on Mean')
plt.ylabel('Frequency')
```

Viewing posterior parameters using prior variables.

```python
posterior(KB18.GP, KB18.A, prior_i)
```

```python
def observed_data_generator(n_try, observed_data):
    result = np.ones(observed_data)
    fails = n_try - observed_data
    result = np.append(result, np.zeros(fails))
    return result
```

```python
en_observed = observed_data_generator(KB18.GP, KB18.A)
```

Running the Metropolis Algorithm using trace

```python
import pymc3 as pm
with pm.Model() as model_1:    
    E_p = pm.Beta('EN_AVG', prior_hits + 1, prior_atbat - prior_hits + 1)
    EN = pm.Bernoulli('EN', p=E_p, observed = en_observed)  
    start = pm.find_MAP()
    trace = pm.sample(10000, start=start)
```

```python
with model_1:
    trace = pm.sample(10000, tune=1000, step=pm.Metropolis())
```

```python
pm.plot_posterior(trace, varnames=['EN_AVG'],ref_val=0.18)
```
